2. Write a Python function called `rentals_month` that retrieves rental data for a given month and year (passed as parameters) from the Sakila database as a Pandas DataFrame.

```python
# let's define a function called 'rentals_month':

def rentals_month (engine,month,year):
    sql = f'''Select *
from sakila.rental
where month(rental_date)={month}
and year(rental_date)={year}'''
    return pd.read_sql_query(sql,engine)

# let's retrieve rental data for may : 

df_may = rentals_month(engine, 5, 2005)
df_may
```

```python
# let's retrieve rental data for june :

df_june = rentals_month(engine, 6, 2005)
df_june
```

3. Develop a Python function called `rental_count_month` that takes the DataFrame provided by `rentals_month` as input along with the month and year and returns a new DataFrame containing the number of rentals made by each customer_id during the selected month and year. 

	The function should also include the month and year as parameters and use them to name the new column according to the month and year, for example, if the input month is 05 and the year is 2005, the column name should be "rentals_05_2005".


	*Hint: Consider making use of pandas [groupby()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html)*

```python
# let's define a function called rental_count_month as stated above :

def rental_count_month(df, month, year):
    # let's group the DataFrame by customer_id and count the rentals
    rental_counts = df.groupby('customer_id').size().reset_index(name=f'rentals_{month}_{year}')

    return rental_counts
```

```python
# df_may = rentals_month(engine, 5, 2005)

# let's call the function 'rental_count_month' to count the number of rentals made per each customer_id during may 2005, as an example :

rental_count_may = rental_count_month(df_may, 5, 2005)
rental_count_may
```

```python
# let's call the function 'rental_count_month' to count the number of rentals made per each customer_id during june 2005, as an example :

rental_count_june = rental_count_month(df_june, 6, 2005)
rental_count_june
```

4. Create a Python function called `compare_rentals` that takes two DataFrames as input containing the number of rentals made by each customer in different months and years. 
The function should return a combined DataFrame with a new 'difference' column, which is the difference between the number of rentals in the two months.

```python
def compare_rentals(df1, df2):
    # let's merge the two DataFrames on customer_id :
    merged_df = pd.merge(df1, df2, on='customer_id', suffixes=('_month1', '_month2'))
    
    # let's calculate the difference between the rental counts of the two months :
    merged_df['difference'] = merged_df.iloc[:, 2] - merged_df.iloc[:, 1]
    
    # let's group by customer_id :
    difference_by_customer_id = merged_df.groupby('customer_id').sum().reset_index()
    
    return difference_by_customer_id


# let's call the function 'compare_rentals' to return the difference in rental counts for each customer_id between May and June :

difference_df_may_june = compare_rentals(rental_count_may, rental_count_june)
difference_df_may_june
```
